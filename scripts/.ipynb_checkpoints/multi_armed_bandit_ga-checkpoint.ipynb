# Algoritmo Genético

```python
def readQAP(name):
    """
    Leer un archivo DAT y lo transforma en un diccionario que contiene el número de nodos (n), matriz de flujo (f) 
    y matriz de distancia (d)

    Args:
        name (str): Nombre del archivo .dat a leer

    Returns:
        instance_dict (dict): Diccionario que contiene el número de nodos (n), matriz de flujo (f) y matriz de distancia (d)
    """
    with open(name, 'r') as file:
        lines = file.readlines()
        
    n_nodos = int(lines[0])
    flow_matrix = np.array([list(map(int, line.split())) for line in lines[1:n_nodos+1]])
    distance_matrix = np.array([list(map(int, line.split())) for line in lines[n_nodos+2:2*n_nodos+2]])
    
    instance_dict = {'n': n_nodos, 'f': flow_matrix, 'd': distance_matrix}
    
    return instance_dict


def evaluateQAP(sol, f, d):
    """
    Evaluar QAP con una solución dada

    Args:
        sol (numpy array): vector con valores de posición para evaluar la función
        f (numpy array): matriz de flujo entre nodos
        d (numpy array): matriz de distancia entre nodos

    Returns:
        Valor de fitness del QAP en base a la solución entregada
    """
    
    fitness = np.sum(f * d[sol[:, None], sol])
    
    return fitness


def generate_population(size, n):
    """
    Genera una población de soluciones aleatorias para el problema QAP

    Args:
        size (int): Tamaño de la población a generar
        n (int): Número de ubicaciones/facilidades en el problema QAP

    Returns:
        population (list): Lista de soluciones aleatorias, donde cada solución es un numpy array 
        que representa una permutación de n elementos
    """
    population = [np.random.permutation(n) for _ in range(size)]
    return population


def fitness(solution, instance):
    """
    Evalúa el costo de una solución al problema QAP

    Args:
        solution (numpy array): Una solución al problema QAP, representada como una permutación de n elementos
        instance (dict): Un diccionario que contiene las matrices de flujo y distancia para una instancia del problema QAP

    Returns:
        cost (int): Costo de la solución evaluada
    """
    cost = evaluateQAP(solution, instance['f'], instance['d'])
    return cost


def crossover(parent1, parent2):
    """
    Aplica operador de cruzamiento entre dos padres

    Args:
        parent1 (numpy array): Primera solución padre
        parent2 (numpy array): Segunda solución padre

    Returns:
        child (numpy array): Hijo resultante de la operación de cruce
    """
    size = len(parent1)
    idx1, idx2 = sorted(np.random.choice(range(size), 2, replace=False))
    child = -np.ones(size, dtype=int)
    child[idx1:idx2] = parent1[idx1:idx2]
    mask = np.isin(parent2, child)
    child[child==-1] = parent2[~mask]
    return child


def selection(population, scores, k=3):
    """
    Selecciona el mejor individuo de k individuos seleccionados aleatoriamente de la población

    Args:
        population (list): Lista de soluciones en la población
        scores (list): Lista de puntajes de fitness correspondientes a las soluciones en la población
        k (int): Número de individuos a seleccionar para la competición

    Returns:
        best_sol (numpy array): La mejor solución de las seleccionadas
    """
    selection_idx = np.random.choice(len(population), k, replace=False)
    best_sol = population[min(selection_idx, key=lambda idx: scores[idx])]
    return best_sol


def mutation(solution):
    """
    Realiza la mutación de intercambio en una solución

    Args:
        solution (numpy array): Solución a mutar

    Returns:
        numpy array: Solución mutada
    """
    idx1, idx2 = np.random.choice(range(len(solution)), 2, replace=False)
    solution[idx1], solution[idx2] = solution[idx2], solution[idx1]
    return solution


def replace_with_new_population(population, new_population, scores):
    """
    Reemplaza la población vieja con la nueva población

    Args:
        population (list): Lista de soluciones en la población antigua
        new_population (list): Lista de soluciones en la nueva población
        scores (list): Lista de puntajes de fitness correspondientes a las soluciones en la población antigua

    Returns:
        tuple: Tupla de dos listas, la nueva población y la lista de costos correspondientes
    """
    population_size = len(population)
    new_scores = [fitness(solution, instance) for solution in new_population]
    population = population + new_population
    scores = scores + new_scores
    idx = np.argsort(scores)[:population_size]
    return [population[i] for i in idx], [scores[i] for i in idx]

def genetic_algorithm(instance, n_population, n_iterations, mutation_rate):
    """
    Implementación del Algoritmo Genético.

    Args:
        instance (dict): Un diccionario que contiene las matrices de flujo y distancia para una instancia del problema QAP
        n_population (int): Tamaño de la población
        n_iterations (int): Número de iteraciones para ejecutar el algoritmo
        mutation_rate (float): Probabilidad de mutación de un individuo en la población

    Returns:
        tuple: Tupla de tres elementos: mejor solución encontrada, costo de la mejor solución, y lista de costo mínimo en cada iteración.
    """
    
    # Se genera la población inicial de acuerdo al tamaño de población
    population = generate_population(n_population, instance['n'])
    
    # Se calcula el costo de cada solución de la población
    costs = [fitness(solution, instance) for solution in population]

    # Se inicializa lista para llevar un registro del costo mínimo en cada iteración
    min_costs = []

    for i in range(n_iterations):
        
        new_population = []
        
        # Se recorre cada solución de la población
        for i in range(n_population):
            parent1 = selection(population, costs)
            parent2 = selection(population, costs)
            child = crossover(parent1, parent2)
            
            if np.random.rand() < mutation_rate:
                child = mutation(child)
                
            new_population.append(child)
            
        population, costs = replace_with_new_population(population, new_population, costs)
        
        # Encontrar el costo mínimo y agregarlo a la lista de seguimiento
        min_cost = min(costs)
        min_costs.append(min_cost)
        
    best_solution = population[np.argmin(costs)]
    best_cost = min_costs

    # Devolver tanto la solución y el puntaje como la lista de seguimiento del costo mínimo
    return best_solution, best_cost, min_costs

def plot_min_scores(min_costs):
    """
    Genera un gráfico de la evolución del costo obtenido tras cada 
    """
    plt.figure(figsize=(10,5))
    plt.plot(min_costs, label='Menor Costo por Iteración')
    plt.xlabel('Iteraciones')
    plt.ylabel('Costo')
    plt.title('Evolución del costo mínimo a través de las iteraciones')
    plt.legend()
    plt.grid(True)
    plt.show()
```

# Multi-Armed Bandits

```python
class MultiArmedBandit:
    """
    Clase que implementa un Bandido Multi-Brazos (Multi-Armed Bandit) utilizando Thompson Sampling.

    Attributes:
        n_arms (int): Número de brazos del bandido.
        alpha (float): Factor de aprendizaje.
        values (numpy array): Array con los valores Q para cada brazo.
        counts (numpy array): Array con el número de veces que cada brazo ha sido seleccionado.
    """
    def __init__(self, n_arms, alpha=0.1):
        """
        Inicializa el Bandido Multi-Brazos.

        Args:
            n_arms (int): Número de brazos del bandido.
            alpha (float, optional): Factor de aprendizaje. Valor por defecto es 0.1.
        """
        self.n_arms = n_arms
        self.alpha = alpha
        self.values = np.zeros(n_arms)  # Inicializa los valores Q con ceros
        self.counts = np.zeros(n_arms)  # Inicializa los conteos con ceros

    def select_arm(self):
        """
        Selecciona un brazo usando Thompson Sampling.

        Returns:
            int: El índice del brazo seleccionado.
        """
        samples = [np.random.beta(self.values[i] + 1, np.maximum(self.counts[i] - self.values[i] + 1, 1e-10)) for i in range(self.n_arms)]
        # Devuelve el brazo con la muestra más alta
        max_arm = np.argmax(samples)
        return max_arm  

    def update(self, chosen_arm, reward):
        """
        Actualiza los valores Q y los conteos con la nueva recompensa.

        Args:
            chosen_arm (int): El índice del brazo seleccionado.
            reward (float): La recompensa recibida después de seleccionar el brazo.
        """
        self.counts[chosen_arm] += 1  # Incrementa el conteo del brazo seleccionado
        q = self.values[chosen_arm]  # Obtiene el valor Q actual del brazo seleccionado
        self.values[chosen_arm] = q + (reward - q) / self.counts[chosen_arm]  # Actualiza el valor Q
```

# Algoritmo Genético con MAB

```python
def genetic_algorithm_mab(instance, n_population, n_iterations, n_arms, alpha_mab, visualization=False):
    """
    Implementación del Algoritmo Genético utilizando un Multi-armed Bandits que selecciona la tasa de mutación de cada
    iteración.

    Args:
        instance (dict): Diccionario que contiene las matrices de flujo y distancia para una instancia del problema QAP.
        n_population (int): Tamaño de la población.
        n_iterations (int): Número de iteraciones para ejecutar el algoritmo.
        n_arms (int): Número de brazos del bandido.
        alpha_mab (float): Factor de aprendizaje del bandido.
        visualization (bool, optional): Indica si se quiere realizar una representación gráfica en tiempo real.

    Returns:
        tuple: Tupla de tres elementos: mejor solución encontrada, costo de la mejor solución, y lista de costos mínimos en cada iteración.
    """
    
    # Se genera la población inicial de acuerdo al tamaño de población
    population = generate_population(n_population, instance['n'])
    
    # Se calcula el costo de cada solución de la población
    costs = [fitness(solution, instance) for solution in population]

    # Inicia un objeto MultiArmedBandit con 10 brazos
    mab = MultiArmedBandit(n_arms=n_arms, alpha=alpha_mab)

    # Se inicializan las listas de seguimiento
    min_costs = []
    mutation_rates = []
    rewards = []
    previous_best_cost = float('inf') 

    for i in range(n_iterations):
        
        # Selecciona un brazo usando el MAB se convierte en un factor de mutación
        chosen_arm = mab.select_arm()
        mutation_rate = 0.1 * (chosen_arm + 1) 

        new_population = []
        for _ in range(n_population):
            parent1 = selection(population, costs)
            parent2 = selection(population, costs)
            child = crossover(parent1, parent2)
            if np.random.rand() < mutation_rate:
                child = mutation(child)
            new_population.append(child)
        population, costs = replace_with_new_population(population, new_population, costs)
        
        # Se encuentra el costo mínimo y se agrega a la lista de seguimiento
        min_cost = min(costs)
        min_costs.append(min_cost)
        
        # Calcula la recompensa como la diferencia entre el previous_best_cost y el min_cost actual
        reward = 0 if i == 0 else previous_best_cost - min_cost
        
        # Se añaden la tasa de mutación y recompensa a su respectiva lista de seguimiento 
        mutation_rates.append(mutation_rate)
        rewards.append(reward)

        # Actualiza el MAB con la recompensa obtenida
        mab.update(chosen_arm, reward)

        # Actualiza el previous_best_score para la próxima iteración
        previous_best_cost = min_cost
        
        # Se genera una gráfico en tiempo real de la evolución de la tasa de mutación y la recompensa en cada iteración
        if visualization:
            # Dibujar la figura
            plt.figure(figsize=(12, 6))

            plt.subplot(2, 1, 1)
            plt.plot(mutation_rates, label='Tasa de Mutación')
            plt.xlabel('Iteraciones')
            plt.ylabel('Tasa de Mutación')
            plt.legend()

            plt.subplot(2, 1, 2)
            plt.plot(rewards, label='Recompensa')
            plt.xlabel('Iteraciones')
            plt.ylabel('Recompensa')
            plt.legend()

            # Para la visualización en tiempo real
            display.clear_output(wait=True)
            display.display(plt.gcf())
            plt.close()
    
    best_solution = population[np.argmin(costs)]
    best_cost = min_cost

    # Devuelve la solución, el costo mínimo y la lista de seguimiento del costo mínimo
    return best_solution, best_cost, min_costs
```

# Comparación

```python
def multiple_runs(genetic_algorithm_func, n_runs, *args, **kwargs):
    # Inicializar una lista para guardar los mejores puntajes de cada ejecución
    best_scores = []

    # Ejecutar el algoritmo genético varias veces
    for _ in range(n_runs):
        print("Ejecución n°", _)
        _, best_score, _ = genetic_algorithm_func(*args, **kwargs)
        best_scores.append(best_score)

    # Devolver los mejores puntajes de cada ejecución y el promedio
    return best_scores
```

## Comparación

```python
# Ejecutar el algoritmo genético sin el bandido multi-brazos
best_scores_no_mab = multiple_runs(genetic_algorithm, 11, instance, n_population, n_iterations, mutation_rate)

# Ejecutar el algoritmo genético con el bandido multi-brazos y diferentes valores de alpha
alphas_mab = [0.1, 0.6, 0.9]
best_scores_mab = {}
for alpha in alphas_mab:
    best_scores_mab[alpha] = multiple_runs(genetic_algorithm_mab, 11, instance, n_population, n_iterations, 10, alpha, visualization=False)

# Calcular la media de los mejores puntajes para cada ejecución
mean_score_no_mab = np.mean(best_scores_no_mab)
mean_scores_mab = {alpha: np.mean(scores) for alpha, scores in best_scores_mab.items()}

print(f"Promedio de mejor puntuación sin MAB: {mean_score_no_mab}")
for alpha, mean_score in mean_scores_mab.items():
    print(f"Promedio de mejor puntuación con MAB (alpha={alpha}): {mean_score}")
```
